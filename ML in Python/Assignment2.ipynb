# Given a decimal integer number, write a function which converts this number into its binary form. Use iteration to solve this problem. E.g., myIntNumber=5 result should be 0101.

```python
#List of numbers
n = [3, 6, 7, 8, 5, 9, 11, 56]
```

```python
def tobinary1(n):
    binary = []
    
    while(n > 0):
        binary.append(n%2)
        n = n // 2
        
    #Just to make sure that all binary digits are of same length
    if len(binary) < 5:
        for i in range(len(binary), 5):
            binary.append(0)
    
    binary.reverse()
    
    
    return binary
```

```python
num = 4
b = tobinary1(num)
print(b)
```

```python
for number in n:
    b = tobinary1(number)
    #Converting into a string element
    string = ' '.join([str(elem) for elem in b])
    print(number, "-->", string)

#Since 56 is greater than 2^5 = 32, it requires more binary digits to be represented
```

# Given an array representing binary form of a decimal integer number. Convert this into corresponding decimal number, e.g., [1,0,0,1] to 9.

```python
def todecimal1(b):
    b.reverse()
    dec = 0
    for i in range(len(b)):
        dec += b[i] * (2 ** i)
    
    return dec
```

```python
#Checking for binary to decimal
for number in n:
    b = tobinary1(number)
    dec = todecimal1(b)
    print(dec)
```

```python
myMedDict={
  "Abelcet":"Aug 1 2016",
"Azithromycin":"Dec 24 2016",
"Arava":"Jan 1 2019",
"Arixtra":"May 31 2022",
"Aplenzin":"Jan 3 2020",
"Antizol":"Aug 31 2023",
"Anadrol-50":"Nov 14 2010"
}

print(myMedDict)
```

```python
from datetime import date, timedelta, datetime
today = date.today()
#changing format to meet dictionary format

Today = today.strftime("%b %d %Y")
print("Today's date = ", Today)
```

```python
def Get_expired_list(myMedDict, Today):
    #Converting into date time object
    date1 = datetime.strptime(Today, '%b %d %Y')
    expired = []
    
    for med, date in myMedDict.items(): 
        #Converting into date time object
        date2 = datetime.strptime(date, '%b %d %Y')
        
        #Getting the difference in number of days
        td = date1 - date2
        #Converting into float value
        days = td / timedelta(days=1)
        
        if days > 0.0:
            expired.append(med)
    
    return expired
```

```python
e = Get_expired_list(myMedDict, Today)
print("List of expired medicines as of Today: \n", e)
```

# Write a recursive algorithm to convert a decimal integer number into its binary form, e.g., myInt=9 to 1001.

```python
def tobinary2(binary, n):
    
    if n > 0:
        #Recursively calling the function
        binary = tobinary2(binary, n//2)
    
    binary.append(n%2)
    #No need to reverse the list because the list is being appended from last element to first element.
    return binary
```

# You have been given a file "crimeData.csv", containing address and number of times a crime has happened at that address. Your task is to write a function which takes a crime name as an input and calculates the mean of that crime, i.e., average number of times that crime has happened at addresses. The function should return all the address (may be in a list) where this crime has occurred more than the mean number of times. 

```python
import pandas as pd
```

```python
#Loading the file into a dataframe
df1 = pd.read_csv('crimeData.csv')
df1.head()
```

```python
#Getting the names of all the columns, hence all the crimes, except the column Address
crimes = df1.columns[1:]
print(crimes)
```

```python
def average(crime, df):
    
    avg = df[crime].mean()
    
    Addr_list = df[df[crime] > avg]['Address']
    
    return Addr_list
```

```python
All_address_list = []

for crime in crimes:
    addr = average(crime, df1)
    print("Crime: ", crime)
    print("\nList of addresses: \n", addr)
    print("-----------------------------------------------------------")
    All_address_list.append(addr)
```

# You have been given a file "timeData2.csv". This file contains date along with the time at which crimes has happened. Your task is to write a function which takes a time interval and tells you which category of crime has occurred the most in that time interval. The time interval can be taken as two separate parameter fromTime and toTime .

```python
#Loading the file into a dataframe
df2 = pd.read_csv('timeData2.csv')
df2.head()
```

```python
#Converting the dates coulmn into datetime format
df2['Dates'] = pd.to_datetime(df2.Dates, format='%m/%d/%Y %H:%M')
```

```python
#Setting a time interval
fromTime = datetime.strptime('12:00', '%H:%M').time()
toTime = datetime.strptime('22:00', '%H:%M').time()
```

```python
def Get_max_crime_category(fromTime, toTime):
    #Getting a list of Crimes committed in the given time interval
    #Assuming global dataframe
    Crimes = df2[df2.Dates.dt.time.between(fromTime, toTime)]['Category']
    
    counts = {}
    #Getting the crimes as keys
    for c in Crimes:
        if c not in counts:
            counts[c] = 0
        counts[c] += 1
        
    if not bool(counts):
        #In case no crime occur in the given interval
        crime = "None"
    else:
        crime = max(counts, key=counts.get)
    
    return crime
```

```python
max_crime = Get_max_crime_category(fromTime, toTime)

print("Crime that occured most in the time interval: ", fromTime, " to ", toTime, "is: ", max_crime)
```
